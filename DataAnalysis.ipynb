- Aykırı değer (outlier) analizi

Aykırı değer analizi yaparken en sık kullandığımız yöntemlerden biri IQR (Interquartile Range) yani çeyrekler arası açıklık yöntemidir. Bu yöntem, verideki olağandışı değerleri belirlemek için kullanılır.

Aşağıdaki görsel, IQR yöntemini ve boxplot’un temel bileşenlerini özetlemektedir:


![IQR](https://www.simplypsychology.org/wp-content/uploads/box-whisker-plot.jpg)


Medyan (Ortanca): Kutunun içindeki orta çizgidir. Verinin ortasındaki değeri gösterir.

Q1 (1. Çeyrek): Verinin %25’lik alt kısmını temsil eder.

Q3 (3. Çeyrek): Verinin %75’lik kısmını kapsayan üst sınırdır.

IQR (Q3 – Q1): Verinin ortadaki %50’lik kısmının yayılımını gösteren aralıktır.

Alt sınır: Q1 – 1.5 × IQR

Üst sınır: Q3 + 1.5 × IQR

Bu sınırların dışında kalan değerler aykırı değer (outlier) olarak değerlendirilir.

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_analysis = df.drop(columns=["id", "Unnamed: 0"], errors="ignore")
numerical_cols = df_analysis.select_dtypes(include=["int64", "float64"]).columns.tolist()

for col in numerical_cols:
    plt.figure(figsize=(7, 1.5))
    sns.boxplot(x=df[col], color="skyblue")
    plt.title(f"{col} - Boxplot")
    plt.show()

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outlier_count = df[(df[col] < lower) | (df[col] > upper)].shape[0]
    print(f"{col} sütununda {outlier_count} aykırı değer bulunuyor.")
```

- Kategorik Değişkenlerin Dağılımı

```python
categorical_cols = df_analysis.select_dtypes(include=["object"]).columns.tolist()

for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df_analysis, x=col)  
    plt.title(f"{col} Değişkeninin Dağılımı")
    plt.xticks(rotation=20)
    plt.tight_layout()
    plt.show()
```

- Sayısal Değişkenlerin Dağılımı

```python
numerical_cols = ["Age", "Flight Distance", "Departure Delay in Minutes", "Arrival Delay in Minutes"]

for col in numerical_cols:
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))

    sns.histplot(data=df, x=col, kde=True, bins=30, color="#ff4d4d", ax=axes[0])
    axes[0].set_title(f"{col} - Histogram ve Yoğunluk Grafiği")
    axes[0].set_ylabel("Count")

    sns.boxplot(data=df, x=col, color="#007acc", ax=axes[1])
    axes[1].set_title(f"{col} - Boxplot")

    plt.tight_layout()
    plt.show()
```

- Hedef Değişken (Satisfaction) ile Kategorik Değişkenler Arasındaki İlişki

```python
kategorik_degiskenler = ['Gender', 'Customer Type', 'Type of Travel', 'Class']
for col in kategorik_degiskenler:
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x=col, hue='satisfaction')
    plt.title(f"{col} ve Memnuniyet İlişkisi")
    plt.xticks(rotation=15)
    plt.tight_layout()
    plt.show()
```

- Hedef Değişken (Satisfaction) ile Sayısal Değişkenler Arasındaki İlişki

```python
numerical_cols = ['Age', 'Flight Distance', 'Departure Delay in Minutes', 'Arrival Delay in Minutes']

for col in numerical_cols:
    plt.figure(figsize=(5.5, 4))
    sns.barplot(data=df, x='satisfaction', y=col, errorbar=None, hue='satisfaction', palette='pastel', legend=False)
    plt.title(f"{col} Ortalaması - Memnuniyete Göre")
    plt.tight_layout()
    plt.show()
```
